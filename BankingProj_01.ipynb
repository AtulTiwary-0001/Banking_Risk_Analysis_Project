```python
from sqlalchemy import create_engine
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

engine = create_engine("mysql+mysqlconnector://root:password@localhost:3306/banking_case")

df = pd.read_sql("SELECT * FROM customer", engine)
```

```python
#  Missing Values
missing = df.isnull().sum()
print('Missing values per column:')
print(missing[missing > 0] if missing.any() else 'No missing values found ✅')
```

```python
# Duplicates
dupes = df.duplicated().sum()
print(f'Duplicate rows: {dupes}')
df.drop_duplicates(inplace=True)
print(f'Shape after dedup: {df.shape}')
```

```python
# Creating Income Band
bins = [0,100000,300000, float('inf')]
labels = ['Low', 'Mid', 'High']

df['income_band'] = pd.cut(df['estimated_income'], bins=bins, labels=labels, right=False)
```

```python
df['income_band'].value_counts().plot(kind= 'bar')
```

```python
# examine the distribution of unique categories in categorical columns
```

```python
categorical_cols = df[[	"nationality", "occupation", "fee_structure", "loyalty_classification", "properties_owned", "risk_weighting", "brid", "gender_id", "iaid","income_band"]].columns

for col in categorical_cols:
    print(f"Value Counts for '{col}':")
    display(df[col].value_counts())
```

```python
# Univariate Analysis
```

```python
for i, predictor in enumerate(df[["nationality", "occupation", "fee_structure", "loyalty_classification", "properties_owned", "risk_weighting", "brid", "gender_id", "iaid","income_band","amount_of_credit_cards" ]].columns
):
    plt.figure(i)
    sns.countplot(data=df, x=predictor)
    
```

```python
# Bivariate Analysis
```

```python
for i, predictor in enumerate(df[["brid", "gender_id", "iaid", "amount_of_credit_cards", "nationality", "occupation", "fee_structure", "loyalty_classification", "properties_owned", "risk_weighting", "income_band"]].columns
):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='nationality')
```

```python
# Histplot of value counts for different occupation
```

```python
for col in categorical_cols:
    if col == "occupation":
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title('Histogram of occupation count')
-    plt.xlabel(col)
    plt.ylabel("count")
    plt.show()
```

```python
 # Numerical Analysis
```

```python
numerical_cols = ['estimated_income', 'superannuation_savings', 'credit_card_balance', 'bank_loans', 'bank_deposits', 'checking_accounts', 'foreign_currency_account', 'business_lending']

# Univariated analysis and visualization
plt.figure(figsize=(15,10))
for i,col in enumerate(numerical_cols):
    plt.subplot(4,3,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
plt.tight_layout() 
plt.show()
```

```python
numerical_cols = ['estimated_income', 'superannuation_savings', 'credit_card_balance', 'bank_loans', 'bank_deposits', 'checking_accounts', 'foreign_currency_account', 'business_lending']

correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot= True, cmap='crest', fmt=".2f")
plt.title("correlation_matrix")
plt.show()
```

```python
numerical_cols = ['estimated_income', 'superannuation_savings', 'credit_card_balance', 'bank_loans', 'bank_deposits', 'checking_accounts', 'foreign_currency_account', 'business_lending']

corr_matrix = df[numerical_cols].corr()

plt.figure(figsize=(16, 13))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, mask=mask, annot=True, fmt='.2f', cmap='RdBu_r',
            center=0, linewidths=0.4, annot_kws={'size': 7},
            cbar_kws={'shrink': 0.7})
plt.title('Full Correlation Matrix — Banking Features', fontweight='bold', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.yticks(fontsize=8)
plt.tight_layout()
plt.show()
```
